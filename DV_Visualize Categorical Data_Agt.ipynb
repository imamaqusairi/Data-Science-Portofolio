# 1. Categorical Scatter plot

## Strip plot

Strip plot looks typically same with scatter plot. What differentiates between two of them is strip plot used for visualizing the categorical data and scatter plot used for visualizing the numerical data.

Basic Syntax : 
`sns.catplot(x="day", y="total_bill", data=tips)`

```python
# practice 1
sns.catplot(x='origin', y='weight', data=df_mpg)
plt.title('Weight Distribution of Each Car Origin')
plt.show()

# practice 3 
sns.catplot(x='origin', y='weight', jitter=False, data=df_mpg) # Unabled the jitter parameter
plt.title('Weight Distribution of Each Car Origin Without Jitter')
plt.show()

# practice 2 
sns.catplot(x='Species', y='SepalWidthCm', data=df_iris)
plt.title('SepaL Widh Distribution of each Species')
plt.show()

# practice 3 
sns.catplot(x='Species', y='SepalWidthCm', jitter=False, data=df_iris) # Unabled the jitter parameter
plt.title('SepaL Widh Distribution of each Species Without Jitter')
plt.show()
```

From graphic above, without jitter parameter makes the strip plot looks neatly and more understandable for readers

## Swarm plot

Same with Strip plot, but swarm plot can help to visualize the categorical visualization without overlapping

Basic Syntax : 
`sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)`

```python
# practice 1 basic 
sns.catplot(x='origin', y='weight', kind='swarm', data=df_mpg)
plt.title('Weight Distribution for Each Origin')
plt.show()

# practice 2 basic
sns.catplot(x='Species', y='PetalWidthCm', kind='swarm', data=df_iris)
plt.title('Petal Width Distribution for Each Species')
plt.show()

# practice 3 setting order 
sns.catplot(x='Species', y='PetalWidthCm', kind='swarm', 
            order=['Iris-setosa', 'Iris-virginica', 'Iris-versicolor'], data=df_iris) # Order parameter help how the categorical plot ordered
plt.title('Petal Width Distribution for Each Species')
plt.show()

# practice 4 create more categorization
sns.catplot(x='origin', y='weight', hue='cylinders', data=df_mpg) # add more categorization using hue parameter 
plt.title('Weight Distribution for Each Origin and Cylinders')
plt.show()
```

# 2. Categorical Distribution plot

## Box plot

Visualization that show information not only data distribution, but alo min, max, median and even outlier.

Basic Syntax :
`sns.catplot(x="day", y="total_bill", kind="box", data=tips)`

```python
# practice 1 basic
sns.catplot(x='origin', y='weight', kind='box', data = df_mpg)
plt.title('Weight Distribution For Each Origin ')
plt.show()

# practice 2 basic
sns.catplot(x='Species', y = 'PetalWidthCm', kind='box', data = df_iris)
plt.title('Petal Width Distribution For Each Species ')
plt.show()

# practice 3 more categorization
sns.catplot(x='origin', y='weight', hue='cylinders', kind='box', data=df_mpg)
plt.title('Weight Distribution for Each Origin and Cylinders')
plt.show()

# practice 4 create boxenplot
sns.catplot(x='origin', y='weight', kind='boxen', data=df_mpg)
plt.title('Weight Distribution For Each Origin using Boxenplots')
```

## Violin plot 

Violin was made by combining box plot and Kernel Density Estimation. 

Basic Syntax :
`sns.catplot(x="total_bill", y="day", hue="sex", kind="violin", data=tips)`

```python
df_iris.head()
```

```python
# practice 1 basic
sns.catplot(x='origin', y='weight', kind='violin', data=df_mpg)
plt.title('Weight Distribution for Each Origin')
plt.show()

# practice 2 basic
sns.catplot(x='Species', y='PetalWidthCm', kind='violin', data=df_iris)
plt.title('Petal Width Distribution for Each Species')
plt.show()

# practice 3 violin plot with showing individual observation
sns.catplot(x='origin', y='weight', kind='violin', inner='stick', data=df_mpg)
plt.title('Weight Distribution for Each Origin')
plt.show()

# practice 4 combine violin plot with swarm or strip plot
g = sns.catplot(x='origin', y='weight', kind='violin', data=df_mpg)
sns.swarmplot(x='origin', y='weight', color='k', size=1.5, data=df_mpg, ax=g.ax)
plt.title('Weight Distribution for Each Origin Combined With Swarm plot')
plt.show()
```

## Bar plot  & Count Plot 

In seaborn modules, using bar plot / count plot help user to visualize not only the categorical data distribution, but also central tendency for each category. 

### Bar Plot 

```python
# practice 1 basic
sns.catplot(x='Species', y='PetalWidthCm', kind='bar', data=df_iris)
plt.title('Basic Bar Plot')
plt.show()

# practice 2 with categorization
sns.catplot(x='origin', y = 'weight', kind = 'bar', hue = 'cylinders', data=df_mpg)
plt.title('Bar Plot With More Categorization')
plt.show()
```

*Notes* : This bar plot used for showing the **mean** of the distribution of a categorical values. Not the **count** of observations of categorical values. 

### Count Plot 

```python
# practice 1
sns.catplot(x='Species', kind ='count', data=df_iris)
plt.title('Basic Vertical Count plot')

# practice 2 
sns.catplot(y='origin', kind='count', data=df_mpg)
plt.title('Basic Horizontal Count plot')
```

We can adapt by changing the visual of the plot vertically or horizontically
