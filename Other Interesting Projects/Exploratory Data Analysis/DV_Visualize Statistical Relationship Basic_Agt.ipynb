### Customized Scatter Plot 

#### Categorized Scatter Plot 

To categorize the data, we can use categorization based on `hue` (color) or `size` (size of dots) parameters in scatter plot 

```python
# Practice 1 visualize relationship between horsepower and weight between number of cylinders
df_iris.head()
sns.relplot(x='horsepower', y='weight', hue='cylinders', data=df_mpg)
```

Seaborn module has a good visualization power that appealing to who see the visualization

```python
# Practice 2 visualize relationship between Sepalwidth and petal length
sns.relplot(x='SepalWidthCm', y='PetalLengthCm', size = 'Species', data=df_iris) # we can use the these
plt.title('Standard Categorized with Size Parameter', fontweight='bold')

sns.relplot(x='SepalWidthCm', y='PetalLengthCm', size = 'Species', sizes=(15,200), data=df_iris) # we can use the these
plt.title('Customized Categorized with Size Parameter', fontweight='bold')
```

#### 2 or More Categorized Scatter Plot 

```python
# Practice 1 visualize relationship between horsepower and weight between number of cylinders and origin of machine
df_iris.head()
sns.relplot(x='horsepower', y='weight', hue='cylinders', style = 'origin', data=df_mpg)
```

```python
# Practice 1 visualize relationship between horsepower and weight between number of cylinders and year of model production
sns.relplot(x='horsepower', y='weight', hue='cylinders', style = 'model_year', data=df_mpg)
```

#### Applying Color Palette

```python
# Practice 1 visualize relationship between horsepower and weight between number of cylinders and year of model production with custom palette
sns.relplot(x='horsepower', y='weight', hue = 'model_year', palette='ch:2,r=.2,l=.6', data=df_mpg)
```

```python
# Practice 2 visualize relationship between Sepalwidth and petal length
sns.relplot(x='SepalWidthCm', y='PetalLengthCm', hue = 'Species', palette = "ch:s=.25,rot=-.25", data=df_iris)
```

## Line Plot 

```python
# load the dataset
df_happy = pd.read_csv('world-happiness-report.csv')
df_happy.head()
```

```python
# for the sake of practice, i will specify on Indonesia country
ind_data = df_happy[df_happy['Country name'] == 'Indonesia']
ind_data.head()
```

### Line Plot Basic 

Basic Syntax : 
`g = sns.relplot(x="time", y="value", kind="line", data=df)`

```python
# practice 1 visualize the GDP trend in Indonesia
sns.relplot(x='year', y='Log GDP per capita', kind='line', data=ind_data)
```

```python
# practice 2 visualize the life expectacy trend at Indonesia 
sns.relplot(x='year', y='Healthy life expectancy at birth', kind='line', data=ind_data)
```
